```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

# punto 1

```python
#extraer información del excel y leer el excel
path = 'city_data.xlsx'
df = pd.read_excel(path)
df
```

```python
def calcular_media(datos):
    return sum(datos) / len(datos) if len(datos) > 0 else 0
print("Media de cada columna")
print(calcular_media(df['GDP (USD Billion)']))
print(calcular_media(df['Population (Millions)']))
print(calcular_media(df['Unemployment Rate (%)']))
print(calcular_media(df['Average Age']))
print(calcular_media(df['Women (%)']))
print(calcular_media(df['Men (%)']))
print(calcular_media(df['Budget (USD Billion)']))
```

```python
def calcular_mediana(datos):
    datos_ordenados = sorted(datos)
    n = len(datos_ordenados)
    if n == 0:
        return None
    if n % 2 == 1:
        return datos_ordenados[n // 2]
    else:
        return (datos_ordenados[n // 2 - 1] + datos_ordenados[n // 2]) / 2
print("Mediana de cada columna") 
print(calcular_mediana(df['GDP (USD Billion)']))
print(calcular_mediana(df['Population (Millions)']))
print(calcular_mediana(df['Unemployment Rate (%)']))
print(calcular_mediana(df['Average Age']))
print(calcular_mediana(df['Women (%)']))
print(calcular_mediana(df['Men (%)']))
print(calcular_mediana(df['Budget (USD Billion)']))
```

```python
def calcular_desviacion_estandar(datos):
    n = len(datos)
    if n == 0:
        return None
    media = calcular_media(datos)
    varianza = sum((x - media) ** 2 for x in datos) / n
    return math.sqrt(varianza)
print("Desviación estandar de cada columna") 
print(calcular_desviacion_estandar(df['GDP (USD Billion)']))
print(calcular_desviacion_estandar(df['Population (Millions)']))
print(calcular_desviacion_estandar(df['Unemployment Rate (%)']))
print(calcular_desviacion_estandar(df['Average Age']))
print(calcular_desviacion_estandar(df['Women (%)']))
print(calcular_desviacion_estandar(df['Men (%)']))
print(calcular_desviacion_estandar(df['Budget (USD Billion)']))
```

```python
def calcular_moda(datos):
    frecuencia = {}
    for dato in datos:
        if dato in frecuencia:
            frecuencia[dato] += 1
        else:
            frecuencia[dato] = 1
    moda = max(frecuencia, key=frecuencia.get)
    repeticiones = frecuencia[moda]
    return moda, repeticiones
print("Moda y repeticiones de cada columna")
print("(moda, repeticiones)")
print(calcular_moda(df['GDP (USD Billion)']))
print(calcular_moda(df['Population (Millions)']))
print(calcular_moda(df['Unemployment Rate (%)']))
print(calcular_moda(df['Average Age']))
print(calcular_moda(df['Women (%)']))
print(calcular_moda(df['Men (%)']))
print(calcular_moda(df['Budget (USD Billion)']))
```

```python
def calcular_covarianza(X1, X2):
    n = len(X1)
    
    if n != len(X2) or n == 0:
        return None  # Verifica que ambas listas tengan el mismo tamaño y no estén vacías
    
    # Calcular las medias
    media_X1 = sum(X1) / n
    media_X2 = sum(X2) / n
    
    # Calcular la covarianza
    covarianza = sum((X1[i] - media_X1) * (X2[i] - media_X2) for i in range(n)) / n
    
    return covarianza
print(calcular_covarianza(df['GDP (USD Billion)'], df['Population (Millions)']))
print(calcular_covarianza(df['Unemployment Rate (%)'], df['Average Age']))
```

# punto 1.4

```python
def calcular_correlacion(X1, X2):
    covarianza = calcular_covarianza(X1, X2)
    desviacion_X1 = calcular_desviacion_estandar(X1)
    desviacion_X2 = calcular_desviacion_estandar(X2)
    
    if desviacion_X1 == 0 or desviacion_X2 == 0:
        return None  # No se puede calcular la correlación si una de las desviaciones estándar es 0
    
    correlacion = covarianza / (desviacion_X1 * desviacion_X2)
    return correlacion
print(calcular_correlacion(df['GDP (USD Billion)'], df['Population (Millions)']))
print(calcular_correlacion(df['Unemployment Rate (%)'], df['Average Age']))
```
