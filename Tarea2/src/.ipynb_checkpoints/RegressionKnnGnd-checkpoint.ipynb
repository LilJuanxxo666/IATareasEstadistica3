# Regresion lineal multiple
* Entrenamiento

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_percentage_error, mean_squared_error

X = df_nuevo.drop(columns=['PUNT_GLOBAL'])
Y = df_nuevo['PUNT_GLOBAL']

#Se define valores más popilares para el train_test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

regression = LinearRegression()
regression.fit(X_train, y_train)

y_pred = regression.predict(X_test)


r = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R2: {r}")
print(f"MAPE: {mape}")
print(f"MSE: {mse}")
```

* Test

```python
df_nuevo_test = df_nuevo_test.reindex(columns=list(X.columns), fill_value=0)

y_pred = regression.predict(df_nuevo_test[list(X.columns)])

test = pd.DataFrame({
    'id': df_test['ESTU_CONSECUTIVO'],
    'GLOBAL_SCORE': y_pred
})

test.to_csv('regression.csv', index=False)
```

# KNN
* Entrenamiento

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

X = df_nuevo.drop(columns=['PUNT_GLOBAL'])
Y = df_nuevo['PUNT_GLOBAL']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Escalar las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear el modelo KNN y ajustarlo 
# 5
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Realizar las predicciones
y_pred = knn.predict(X_test_scaled)

# Calcular el MSE
mse = mean_squared_error(y_test, y_pred)
print(f"El MSE para 5 vecinos cercanos es {mse}")

# 10
# Crear el modelo KNN y ajustarlo
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train_scaled, y_train)

# Realizar las predicciones
y_pred = knn.predict(X_test_scaled)

# Calcular el MSE
mse = mean_squared_error(y_test, y_pred)
print(f"El MSE para 10 vecinos cercanos es {mse}")

# 20
# Crear el modelo KNN y ajustarlo
knn = KNeighborsRegressor(n_neighbors=20)
knn.fit(X_train_scaled, y_train)

# Realizar las predicciones
y_pred = knn.predict(X_test_scaled)

# Calcular el MSE
mse = mean_squared_error(y_test, y_pred)
print(f"El MSE para 20 vecinos cercanos es {mse}")

# 30
# Crear el modelo KNN y ajustarlo
knn = KNeighborsRegressor(n_neighbors=30)
knn.fit(X_train_scaled, y_train)

# Realizar las predicciones
y_pred = knn.predict(X_test_scaled)

# Calcular el MSE
mse = mean_squared_error(y_test, y_pred)
print(f"El MSE para 30 vecinos cercanos es {mse}")

print(f"El mse de 30, es el mejor con un mse de {mse}")
```

* test

```python
X_test_real = df_nuevo_test[list(X.columns)]
X_test_real_scaled = scaler.transform(X_test_real)

knn = KNeighborsRegressor(n_neighbors=30)
knn.fit(X_train_scaled, y_train)

y_pred = knn.predict(X_test_real_scaled)

test = pd.DataFrame({
    'id': df_test['ESTU_CONSECUTIVO'],
    'GLOBAL_SCORE': y_pred
})

test.to_csv('knn.csv', index=False)
```

# GBM
* Entrenamiento

```python
from sklearn.ensemble import GradientBoostingRegressor

# Crear el modelo Gradient Boosting Regressor con parámetros iniciales
gbm = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Ajustar (entrenar) el modelo con los datos de entrenamiento escalados
gbm.fit(X_train_scaled, y_train)

# Realizar predicciones sobre los datos de entrenamiento y prueba
y_pred_train = gbm.predict(X_train_scaled)
y_pred_test = gbm.predict(X_test_scaled)

# Calcular el MSE (Error Cuadrático Medio) y el MAPE (Error Absoluto Medio Porcentual) para entrenamiento y prueba
mse_train = mean_squared_error(y_train, y_pred_train)
mape_train = mean_absolute_percentage_error(y_train, y_pred_train)

mse_test = mean_squared_error(y_test, y_pred_test)
mape_test = mean_absolute_percentage_error(y_test, y_pred_test)

# Imprimir los resultados
print(f"MSE en entrenamiento: {mse_train}")
print(f"MAPE en entrenamiento: {mape_train}")
print(f"MSE en prueba: {mse_test}")
print(f"MAPE en prueba: {mape_test}")
```

* Test

```python
y_pred = gbm.predict(X_test_real_scaled)

test = pd.DataFrame({
    'id': df_test['ESTU_CONSECUTIVO'],
    'GLOBAL_SCORE': y_pred
})

test.to_csv('gbm.csv', index=False)
```
